```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./fox.png")
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

# image resizing


```python
print(img_rgb.shape)
```

```python
newHeight,newWidth=400,400
resized_img=cv2.resize(img_rgb,(newWidth,newHeight))
plt.imshow(resized_img)
```

# image Scaling

```python
scale_percent=50
sHeight=int(img_rgb.shape[0]*scale_percent/100)
sWidth=int(img_rgb.shape[1]*scale_percent/100)
img_scaled=cv2.resize(img_rgb,(sWidth,sHeight),interpolation=cv2.INTER_AREA)
print(img_scaled.shape)
plt.imshow(img_scaled)
```

# interplation

```python
fig,axs=plt.subplots(1,5)
re_width=600
re_height=600
img_rgb_resized=cv2.resize(img_rgb,(100,100))
methods={"nearest":cv2.INTER_NEAREST
,"linear": cv2.INTER_LINEAR
,"area": cv2.INTER_AREA
,"cubic":cv2.INTER_CUBIC
,"lanc":cv2.INTER_LANCZOS4
}
resized_images={}
for name, method in methods.items():
    resized_images[name]=cv2.resize(img_rgb_resized,(re_width,re_height),interpolation=method)
    cv2.imwrite(f'./{name}.jpg',resized_images[name])

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.imshow(img_rgb_resized)
plt.title("original")
plt.axis("off")

for i,(name,resized_img) in enumerate(resized_images.items(),start=2):
    plt.subplot(2,3,i)
    plt.imshow(resized_img)
    plt.title(name)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
