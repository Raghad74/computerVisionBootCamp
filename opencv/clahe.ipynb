```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./images/fox.png")
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_grayscale=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
equalized_img=cv2.equalizeHist(img_grayscale)
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(4,4))
clahe_equalize_img=clahe.apply(img_grayscale)

plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
plt.imshow(img_grayscale,cmap="gray")
plt.title("og")

plt.subplot(1,3,2)
plt.imshow(equalized_img,cmap="gray")
plt.title("histogram equalization")

plt.subplot(1,3,3)
plt.imshow(clahe_equalize_img,cmap='gray')
plt.title("clahe")

plt.show()
```

```python
img_yuv=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
img_yuv[:,:,0]=clahe.apply(img_yuv[:,:,0])

img_yuv_bgr=cv2.cvtColor(img_yuv,cv2.COLOR_YUV2BGR)
img_clahe_rgb=cv2.cvtColor(img_yuv_bgr,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(img_clahe_rgb)
plt.title("clahe equalization")

plt.show()
```
