```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
height,width=500,700
color=(128,127,0)
canvas=np.full((height,width,3),color,dtype=np.uint8)
canvas_rgb=cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
plt.imshow(canvas_rgb)
```

# drawing a polygon

```python
points=np.array([[150,400],[300,400],[225,300]],np.int32)
reshaped_points=points.reshape((-1,1,2))
print(reshaped_points)
cv2.polylines(canvas_rgb,[reshaped_points],True,(255,255,255),3)
plt.imshow(canvas_rgb)
```
