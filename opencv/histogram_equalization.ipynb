```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./images/fox.png")
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_grayscale=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
og_hist,_=np.histogram(img_grayscale.flatten(),256,[0,256])
equalized_img=cv2.equalizeHist(img_grayscale)
equalized_hist,_=np.histogram(equalized_img.flatten(),256,[0,256])

plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.imshow(img_grayscale,cmap="gray")
plt.title("og image")

plt.subplot(2,2,2)
plt.plot(og_hist,color='blue')
plt.title("og histogram")

plt.subplot(2,2,3)
plt.imshow(equalized_img,cmap="gray")
plt.title("equalized image")

plt.subplot(2,2,4)
plt.plot(equalized_hist,color='blue')
plt.title("equalized histogram")

plt.tight_layout()
plt.show()

```

```python
img_yuv=cv2.cvtColor(img,cv2.COLOR_BGR2YUV)
img_yuv[:,:,0]=cv2.equalizeHist(img_yuv[:,:,0])
equalized_img_bgr=cv2.cvtColor(img_yuv,cv2.COLOR_YUV2BGR)
equalized_img_rgb=cv2.cvtColor(equalized_img_bgr,cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(equalized_img_rgb)
plt.title("equalized")

plt.tight_layout()
plt.show()
```
