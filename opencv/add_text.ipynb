```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
height,width=400,600
color=(255,127,0)
canvas=np.full((height,width,3),color,np.uint8)
canvas=cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB)
plt.imshow(canvas)
```

```python
text="Hi there opencv"
font=cv2.FONT_HERSHEY_COMPLEX
font_scale=1.5
org=(50,100)
color=(255,255,255)
thickness=2
cv2.putText(canvas,text,org,font,font_scale,color,thickness,cv2.LINE_AA)
plt.imshow(canvas)
```
