```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./images/fox.png",cv2.IMREAD_GRAYSCALE)
```

```python
#sobel
sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
combined_sobel=cv2.magnitude(sobel_x,sobel_y)

#laplacian
laplacian=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
laplacian_abs=cv2.convertScaleAbs(laplacian)

#canny
canny=cv2.Canny(img,180,200)

titles=["og","sobel","laplacian","canny"]
images=[img,combined_sobel,laplacian_abs,canny]

plt.figure(figsize=(12,6))
for i,(name,image) in enumerate(zip(titles,images),start=1):
    plt.subplot(2,2,i)
    plt.imshow(image,cmap="gray")
    plt.title(name)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
