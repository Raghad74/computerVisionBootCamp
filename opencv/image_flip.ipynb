```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./images/fox.png")
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

# flip

```python
#1-> horizontally
#0->vertically
#-1-> both
flip_horizon=cv2.flip(img_rgb,1)
flip_vertical=cv2.flip(img_rgb,0)
flip_bothAx=cv2.flip(img_rgb,-1)
titles=["og","horiz","vertical","both"]
images=[img_rgb,flip_horizon,flip_vertical,flip_bothAx]
plt.figure(figsize=(12,6))
for i,(title,image) in enumerate(zip(titles,images),start=1):
    plt.subplot(1,4,i)
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
plt.tight_layout()
plt.show()
```

# Rotate

```python
rotate_90_clockwise=cv2.rotate(img_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterClockwise=cv2.rotate(img_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180=cv2.rotate(img_rgb,cv2.ROTATE_180)
titles=["og","90 clockwise","90 counter","180"]
images=[img_rgb,rotate_90_clockwise,rotate_90_counterClockwise,rotate_180]
plt.figure(figsize=(12,6))
for i,(title,image) in enumerate(zip(titles,images),start=1):
    plt.subplot(1,4,i)
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
plt.tight_layout()
plt.show()
```
