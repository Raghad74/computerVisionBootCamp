```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./images/fox.png")
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
channels=cv2.split(img_rgb)
colors=['red','green','blue']
for colour, channel in zip(colors,channels):
    hist=cv2.calcHist([channel],[0],None,[256],[0,256])
    plt.plot(hist,color=colour)
plt.title('color histogram')
plt.xlabel('Pixel Intencity')
plt.ylabel('Frequency')
plt.show()
```

```python
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(img_hsv)

h_hist=cv2.calcHist([h],[0],None,[180],[0,180])
s_hist=cv2.calcHist([s],[0],None,[256],[0,256])
v_hist=cv2.calcHist([v],[0],None,[256],[0,256])

fig,axes=plt.subplots(1,3,figsize=(15,5))

axes[0].plot(h_hist,color='orange')
axes[0].set_title('hue hist')
axes[0].set_xlabel("bins")
axes[0].set_ylabel("frequency")

axes[1].plot(s_hist,color='green')
axes[1].set_title('staturation hist')
axes[1].set_xlabel("bins")
axes[1].set_ylabel("frequency")

axes[2].plot(v_hist,color='blue')
axes[2].set_title('value hist')
axes[2].set_xlabel("bins")
axes[2].set_ylabel("frequency")

plt.tight_layout()
plt.show()


```
