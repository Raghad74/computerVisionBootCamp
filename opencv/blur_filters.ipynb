```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./images/fox.png")
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
kernelS=5
average_blur=cv2.blur(img_gray,(kernelS,kernelS))
plt.imshow(average_blur,cmap='gray')
```

```python
guassian_blur=cv2.GaussianBlur(img_gray,(kernelS,kernelS),0)
plt.imshow(guassian_blur,cmap='gray')
```

```python
median_blur=cv2.medianBlur(img_gray,3)
plt.imshow(median_blur,cmap='gray')
```
