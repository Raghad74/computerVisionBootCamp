```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./images/fox.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
rows,cols,_=img.shape
```

```python
input_points=np.float32([[50,50],[cols-50,50],[50,rows-50],[cols-50,rows-50]])
output_points=np.float32([[10,100],[cols-100,50],[100,rows-50],[cols-50,rows-100]])
M=cv2.getPerspectiveTransform(input_points,output_points)
perspective_img=cv2.warpPerspective(img,M,(cols,rows))

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(perspective_img)
plt.title("perspictive")

plt.tight_layout()
plt.show()


```

# affine

```python
input_points=np.float32([[50,50],[cols-50,50],[50,rows-50]])
output_points=np.float32([[10,100],[cols-100,50],[100,rows-50]])
M=cv2.getAffineTransform(input_points,output_points)
perspective_img=cv2.warpAffine(img,M,(cols,rows))

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(perspective_img)
plt.title("affine")

plt.tight_layout()
plt.show()
```

# Rotation

```python
M=cv2.getRotationMatrix2D([rows//2,cols//2],45,1)
perspective_img=cv2.warpAffine(img,M,(cols,rows))

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(perspective_img)
plt.title("affine rotated")

plt.tight_layout()
plt.show()
```

```python
shear_x,shear_y=0,0.5
M=np.float32([[1,shear_x,0],[shear_y,1,0]])
perspective_img=cv2.warpAffine(img,M,(cols+int(shear_x*rows),rows+int(shear_y*cols)))

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(perspective_img)
plt.title("affine shear")

plt.tight_layout()
plt.show()
```

# Translation

```python
t_x,t_y=50,100
M=np.float32([[1,0,t_x],[0,1,t_y]])
perspective_img=cv2.warpAffine(img,M,(cols+t_y,rows+t_x))

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("og")

plt.subplot(1,2,2)
plt.imshow(perspective_img)
plt.title("affine translation")

plt.tight_layout()
plt.show()
```
