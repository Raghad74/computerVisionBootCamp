```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("./fox.png")
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_rgb)
```

```python
lower_white=np.array([0,0,190])
upper_white=np.array([30,80,255])
```

```python
mask=cv2.inRange(img_hsv,lower_white,upper_white)
```

```python
plt.imshow(mask)
```

```python
black_hsv=np.uint8([[[120,255,0]]])
black_bgr=cv2.cvtColor(black_hsv,cv2.COLOR_HSV2BGR)
```

```python
img_res=img.copy()
img_res[mask>0]=black_bgr
```

```python
img_res_rgb=cv2.cvtColor(img_res,cv2.COLOR_BGR2RGB)
```

```python
fig, axs= plt.subplots(1,3,figsize=(12,6))
axs[0].imshow(img_rgb)
axs[0].set_title("original")
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title("mask")
axs[1].axis('off')


axs[2].imshow(img_res_rgb)
axs[2].set_title("result")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
