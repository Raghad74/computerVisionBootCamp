```python
import torch
```

```python
#by default rand gives values between 0 and 1
random_tensor=torch.rand(5)
print(random_tensor)
print(random_tensor.ndim)
```

```python
random_tensor=torch.rand(2,5,3)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
torch.manual_seed(3)
random_tensor=torch.rand(5)
print(random_tensor)
```

generate random integers

```python
random_integers=torch.randint(0,10,(2,5,3))
print(random_integers)
```

```python
#generate random numbers from a normal distribution
#torch.normal(mean,std,size)
random_numbers_normal=torch.normal(0.5,2,(5,))
print(random_numbers_normal)
```

## creating noise image using pytorch

```python
import matplotlib.pyplot as plt
```

```python
#h,w of image 256x256
uniform_noise=torch.rand((256,256))
plt.imshow(uniform_noise.numpy(),cmap='gray')
plt.title('uniform noise img')
plt.axis('off')
plt.show
```

```python
#h,w,c of image  256x256x3
uniform_noise_rgb=torch.rand((256,256,3))
plt.imshow(uniform_noise_rgb.numpy())
plt.title('uniform noise img')
plt.axis('off')
plt.show
```

```python
print("shape", uniform_noise_rgb.shape)
print("shape", uniform_noise_rgb.dtype)
print("Min value",torch.min(uniform_noise_rgb))
print("Max value",torch.max(uniform_noise_rgb))
```
